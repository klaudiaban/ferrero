```python
import pandas as pd
```

```python
df = pd.read_csv('lista_zawiadomień.csv', encoding='utf-8-sig', sep=";")
df.head(50)
```

```python
missing_summary = df.isnull().sum().to_frame(name='Missing Values')
missing_summary['% Missing'] = (missing_summary['Missing Values'] / len(df)) * 100

missing_summary
```

```python
# Handle missing values in Nr zlec., Opis i Priorytet
df['Nr zlec.'] = df['Nr zlec.'].fillna('Brak danych')
df['Opis'] = df['Opis'].fillna('Brak opisu')
df['Priorytet'] = df['Priorytet'].fillna('Nieokreślony')

# Delete rows with NaN values in Oznaczenie (name of the machine)
df = df.dropna(subset=['Oznaczenie'])

df
```

```python
# Delete not needed columns: Utworzone przez, Opis and Status systemu
df = df.drop(columns=['Utworzone przez', 'Opis', 'Status systemu'])

df
```

```python
# Change type of the column Data zawiadom. to date
df['Data zawiadom.'] = pd.to_datetime(df['Data zawiadom.'], format='%d.%m.%Y', errors='coerce')

df
```

```python
# The number of unique values in Nr zlec. and Oznaczenie
unique_counts = {
    'Nr zlec.': df['Nr zlec.'].nunique(),
    'Oznaczenie': df['Oznaczenie'].nunique()
}

unique_counts
```

```python
# Display top 50 most common words in Oznaczenie
from collections import Counter
import re

all_words = ' '.join(df['Oznaczenie'].astype(str)).lower()
words = re.findall(r'\b\w{3,}\b', all_words)  

word_counts = Counter(words)
common_words = word_counts.most_common(50)

import pandas as pd
common_words_df = pd.DataFrame(common_words, columns=['Word', 'Frequency'])

common_words_df
```

```python
# Display top 50 most common bigrams (2-word phrases) in Oznaczenie
from collections import Counter

phrases = []
for line in df['Oznaczenie'].astype(str).str.lower():
    words = re.findall(r'\b\w{3,}\b', line)
    phrases.extend([f"{words[i]} {words[i+1]}" for i in range(len(words)-1)])

bigram_counts = Counter(phrases)
common_bigrams = bigram_counts.most_common(50)

common_bigrams_df = pd.DataFrame(common_bigrams, columns=['Phrase', 'Frequency'])

common_bigrams_df
```

```python
mask = df['Oznaczenie'].str.lower().str.contains('kartoniarek')
unique_machines = df.loc[mask, 'Oznaczenie'].unique()

sum = 0
machines = []
for machine in unique_machines:
    sum = sum + 1
    machines.append(machine)

print(machines)
sum
```
