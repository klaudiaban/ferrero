```python
import pandas as pd
```

```python
df = pd.read_csv('lista_zawiadomień.csv', encoding='iso-8859-2', sep=";")
df.head(50)
```

```python
replacements = {
    'ó': ['?ó', '?o'],
    'ł': ['?l', '?'],
    'ś': ['?s'],
    'ń': ['?n'],
    'ć': ['?c'],
    'ż': ['?z'],
    'ź': ['?x', '?x'],
    'ą': ['?a'],
    'ę': ['?e'],
    'Ł': ['?L'],
    'Ś': ['?S'],
    'Ń': ['?N'],
    'Ć': ['?C'],
    'Ż': ['?Z'],
    'Ź': ['?X'],
    'Ą': ['?A'],
    'Ę': ['?E']
}

def clean_polish_chars(text):
    if not isinstance(text, str):
        return text
    for correct_char, corrupted_list in replacements.items():
        for corrupted in corrupted_list:
            text = text.replace(corrupted, correct_char)
    return text

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(clean_polish_chars)

df.head(50)
```
